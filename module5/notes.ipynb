```python
flight_delay_df = pd.read_csv('../data/flight_delays.csv',index_col='Month')

flight_delay_df.head()
```

```python
plt.figure(figsize=(16,6))

plt.title( 'Average Delay for Spirit Airlines')

sns.barplot(x=flight_delay_df.index,y=flight_delay_df['NK'])

plt.ylabel('Arrival Delay in Minutes')

```

```python
plt.figure(figsize=(16,6))

plt.title( 'Average Delay for Each Airline, by Month')

sns.heatmap(data=flight_delay_df, annot=True)
```
