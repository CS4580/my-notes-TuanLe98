```python
import pandas as pd

data_df = pd.read_csv('../data/hotel_bookings.csv')

data_df.head()
```

```python
# see dataset dtype

#data_df.dtypes

# Task: Create a list of those columns/features that are not type 'object'
list_objects= list(data_df.select_dtypes(['object']).columns)

list_objects


#cat_cols = [col for col in data_df.columns if data_df[col].dtype == 'object']



```

```python
#Task: Create a new dataframe from only those columns

cat_df =  data_df[list_objects]

cat_df.head()
```

```python
#Task: print unique values of each feature 

for col in cat_df.columns:
    print(f'{col}:\n{cat_df[col].unique()}')
```

```python
#Task: Cast the categorical feature  to the category date type for 'arrival_date_month'

month_map = {'January':1,'February':2,'March':3,'April':4,'May':5,'June':6,'July':7,'August':8,'September':9,'October':10,'November':11,'December':12}
#cat_df['arrival_date_month']=cat_df['arrival_date_month'].map(month_map)

cat_df.loc[:,'arrival_date_month']=cat_df['arrival_date_month'].map(month_map)

cat_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder


le = LabelEncoder()

```

```python
#cat_df['country'] = le.fit_transform(cat_df['country'])
cat_df.loc[:,'country'] = le.fit_transform(cat_df['country'])

cat_df.head()
```

```python
processed_col=[['arrival_date_month', 'country']]
one_hot_cols = [col for col in cat_df.columns if col not in processed_col]

one_hot_cols
```

```python
#One-Hot Encoding


dummy_df = pd.get_dummies(data=cat_df,columns=one_hot_cols)

dummy_df.head()

```

```python
# merge back to the cat_df

cat_df =  pd.concat([cat_df,dummy_df],axis=1)

cat_df
```

```python
num_df =  data_df.drop(columns=list_objects,axis=1)

num_df.head()
```

```python
final_df = pd.concat([num_df,cat_df],axis=1)
```

```python
final_df.to_csv('../data/hotel_booking_1.csv')
```
