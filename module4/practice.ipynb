```python
import pandas as pd

reviews_df = pd.read_csv('../data/winemag-data-130k-v2.csv')
```

```python
reviews_df.head()
```

```python
# Task: who are the most common wine reviewer.
# create a Series whose index is the 'taster_twitter_handle'
# Whose values count now many reviews each person wrote

reviewer = reviews_df.groupby('taster_twitter_handle').count()
# reviewer = reviews_df.groupby('taster_twitter_handle')[taster_twitter_handle].count()
reviewer
```

```python
# Task 2: What is the best wine i can buy for a given amount for money?
# Create a Series whose index is the wine price and whose values are the max number of points
# a wine costing that much was given in a review
# sort the series by price, ascending ($4 at the top and 3300 at the bottom)

price_review = reviews_df.groupby('price')['points'].max().sort_index()

price_review
```

```python
# Task: what are the minimum and the maximum prices for each variety of wine 

# create a dataframe whose index is variety and whose values are min max value

variety_review = reviews_df.groupby(['variety']).price.agg([min, max])

variety_review
```

```python
# Task: what are the most expensive wine varieties?
# create a variable 'sorted_varieties' containing a copy of the DataFrame from the previous task.

sorted_varieties = variety_review.sort_values(by=['min','max'], ascending=False)

sorted_varieties
```

```python
# Task: Create a series whose index is 'reviewer' and whose values is the average review score given by the reviewer.
# hintL you need 'taster name' and 'points'


reviewer_points = reviews_df.groupby('taster_name')['points'].mean().sort_index()

reviewer_points
```
